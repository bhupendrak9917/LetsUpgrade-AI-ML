```python
df8 = df1[['Age', 'DistanceFromHome', 'Education', 'JobLevel', 'MonthlyIncome', 'NumCompaniesWorked', 'PercentSalaryHike',
          'TotalWorkingYears', 'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
```

```python
fig = plt.figure(figsize=(25,15))
for i, col in enumerate(df8.columns):
    plt.subplot(3, 4, i+1)
    plt.boxplot(df8[col])
    plt.title(col)

plt.show()
```

```python
fig = plt.figure(figsize=(25,15))
for i, col in enumerate(df8.columns):
    plt.subplot(3, 4, i+1)
    plt.hist(df8[col])
    plt.title(col)

plt.show()
```
